<h3><b>Configurações Iniciais:</b></h3>

```python
import pandas as pd

df = pd.read_excel('/content/sample_data/default_of_credit_card_clients__courseware_version_1_21_19.xls')
```

<h3><b>Exercício 1 (2,0pts):</b> Crie listas com nomes para as características financeiras restantes.</h3>

```python
colunas_faturas = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3',
                 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']

colunas_pagamentos = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3',
                'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']

print("Colunas Faturas: ", colunas_faturas, "\nColunas Pagamentos: ", colunas_pagamentos)
```

<h3><b>Exercício 2 (3,0pts):</b> Use .describe() para examinar as sínteses estatísticas das características de valor da fatura. Reflita sobre o que viu. Faz sentido?</h3>
<p><b>Resposta: </b>É possível verificar valores muito altos no máximo e média com mediana menor. O que indica uma distribuição assimétrica. Isso faz sentido, pois alguns clientes podem ter dividas bem maiores que a maioria.</p>.

```python
df[colunas_faturas].describe()
```

<h3><b>Exercício 3 (4,0pts):</b> Visualize as características de valor da fatura usando uma grade 2 por 3 de plotagens de histograma. Dica: Você pode usar 20 bins para essa visualização.<h3>

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 3, figsize=(15, 8))
axs = axs.flatten()

for i, col in enumerate(colunas_faturas):
    axs[i].hist(df[col], bins=20, edgecolor='black')
    axs[i].set_title(col)

plt.tight_layout()
plt.show()
```

<h3><b>Exercício 4 (4,0pts):</b> Obtenha o resumo de .describe() para as características de valor do pagamento. Faz sentido?</h3>

<h3><b>Exercício 5 (4,0pts):</b> Plote um histograma das características de pagamento da fatura semelhante ao das características de valor da fatura, mas aplique também alguma rotação aos rótulos do eixo x com o argumento de palavra-chave xrot para que eles não se sobreponham. <h3><b>Exercício 7 (4,0pts):</b> Ignorando os pagamentos iguais a 0 usando a máscara que criou na etapa anterior, utilize o método .apply() do pandas e o método np.log10() do NumPy para plotar histogramas de transformações logarítmicas dos pagamentos diferentes de zero. Considere os resultados. Dica: Você pode usar .apply() para aplicar qualquer função, inclusive log10, a todos os elementos de um DataFrame ou de uma coluna usando a sintaxe a seguir: .apply(<nome_função>).</h3>

```python
import numpy as np

non_zero_df = df[colunas_pagamentos].replace(0, np.nan)
log_df = non_zero_df.apply(np.log10)

fig, axs = plt.subplots(2, 3, figsize=(15, 8))
axs = axs.flatten()

for i, col in enumerate(colunas_pagamentos):
    axs[i].hist(log_df[col].dropna(), bins=20, color='orange', edgecolor='black')
    axs[i].set_title(f'log10({col})')

plt.tight_layout()
plt.show()
```
